```python
import numpy as np
import overreact as rx
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.set(style="white", context="notebook", palette="colorblind", font_scale=1.1)
```

```python
values = [9.0e7, 7.0e7, 9.7e7, 2.3e7, 1.1e8]
sources = ["Neta et al.", "Men'kin et al.", "Hickel & Sehested", "Dzib et al.", "Present work"]
```

```python
n = 3
plt.bar(range(n), np.log10(values[:n]),
        color=sns.color_palette("colorblind")[0], align="center")
plt.bar(range(n, len(sources)), np.log10(values[n:]),
        color=sns.color_palette("colorblind")[2], align="center")
plt.xticks(0.5 + np.arange(len(sources)), sources, rotation=-15)
plt.gca().get_xticklabels()[-1].set_weight("bold")
plt.ylim((7.25, 8.125))
plt.xlim((-0.5, len(sources) - 0.5))
plt.ylabel("log$_{10}$(k (M$^{-1}$s$^{-1}$))")
# plt.ylabel("k (M$^{-1}$s$^{-1}$)")
plt.hlines(np.log10(np.mean(values)), -1, len(sources),
           color=sns.color_palette("colorblind")[1], label="avg. exp.")
plt.legend()
```
